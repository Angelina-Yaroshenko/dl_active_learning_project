```python
#file = open('../data/conll2003/conll2003/eng.testa.txt','r')
file = open('../data/conll2003/conll2003/eng.testb.txt','r')
#file = open('../data/conll2003/conll2003/eng.train.txt','r')
txt = file.read()
txt = txt.split('\n')
#txt = [x for x in txt if x != '-DOCSTART- -X- O O']
```

```python
def appender(text_list):
    sentence = ' '.join(word for word in text_list)
    return sentence
text_list=[]
tag0_list =[]
tag1_list =[]
tag2_list =[]

corpus = []
tag0 =[]
tag1 = []
tag2 = []

for i in range(len(txt)):
    words = txt[i].split(' ')
    if txt[i] != '':
        #print(words[0],words[1],words[2],words[3])
        text_list.append(words[0])
        tag0_list.append(words[1])
        tag1_list.append(words[2])
        tag2_list.append(words[3])
    if txt[i] == '':
        corpus.append(appender(text_list))
        text_list = []
        tag0.append(appender(tag0_list))
        tag0_list = []
        tag1.append(appender(tag1_list))
        tag1_list =[]
        tag2.append(appender(tag2_list))
        tag2_list = []
```

```python
entitylist = ['I-ORG','I-MISC','I-PER']
```

```python
def numbers(tag2,criteria):
    rez = []
    i = 0
    mass=[]
    for element in tag2:
        test = element.split(' ')
        for element2 in test:
            if element2 == criteria:
                mass.append(i)
            i+=1
        i = 0
        rez.append(mass)
        mass = []
    return(rez)
something = numbers(tag2,'I-ORG')
#print(something)
```

```python
def spander(j,number):
    mas = []
    i = 0
    kon_v = len(corpus[j].split(' ')[number])
    for idx, item in enumerate(corpus[j]):
        if number == 0:
            mas.append(0)
            mas.append(kon_v)
            return mas
        if item == ' ':
            i += 1
        if number == i:
            mas.append(idx+1)
            mas.append(idx+kon_v+1)
            return(mas)
#print(spander(3,14))
```

```python
#print(something)
def final_span(Tag):
    List1 = []
    mas1 = []
    for idx, element in enumerate(numbers(tag2,Tag)):
        #print(element)
        if element == []:
            List1.append([])
        if len(element) == 1:
            List1.append([spander(idx,element[0])])
        if len(element) > 1:
            for i in element:
                mas1.append((spander(idx,i)))
            List1.append(mas1)
            mas1 = []
    return List1
#print(List1)       
```

```python
List1 = final_span('I-MISC')
List2 = final_span('I-ORG')
List3 = final_span('I-PER')
```

```python
data =  {'Corpus':corpus,'Tag':tag2,'I-ORG':List2,'I-MISC':List1,'I-PER':List3}
df = pd.DataFrame(data) 
df.head()
```
